```python
#接近中心性排序
ccenter = sorted(b.items(), key=lambda x: x[1], reverse=True)
print(ccenter)
```

```python
#计算排序位置
def getIndex(a, b) :
    count = 0
    for x in a:
        count = count + 1
        if x[0] == b:
            break
    return count

print(getIndex(ccenter, '0'))
```

```python
import networkx as nx
import xlwt
###计算统计指标

#计算介数中心度
print("Betweenness centrality")
b = nx.betweenness_centrality(G)
for v in G.nodes():
    print(v, b[v])
#计算紧密中心度
print("Closeness centrality")
c = nx.closeness_centrality(G)
for v in G.nodes():
    print(v, c[v])
#计算度中心度
print("Degree centrality")
d = nx.degree_centrality(G)
for v in G.nodes():
    print(v, d[v])


import xlwt
 ##导出到excel表格中
    
#创建Workbook,相当于创建Excel
xls = xlwt.Workbook(encoding='utf-8')
#创建sheet，Sheet1为表的名字，cell_overwrite_ok为是否覆盖单元格
sheet = xls.add_sheet('sheet1', cell_overwrite_ok=True)
# 创建的文件夹，用来写入处理后的数据


file = "D:\\facebook_new.xls"
    #向表中添加数据
sheet.write(0, 0,'id')
sheet.write(0, 1, "Betweenness centrality")
sheet.write(0, 2, "Closeness centrality")
sheet.write(0, 3, "Degree centrality")
m = 1
for v in G.nodes():
            sheet.write(m,0,v)
            sheet.write(m, 1, b[v])
            sheet.write(m, 2, c[v])
            sheet.write(m, 3, d[v])
            m = m + 1
# 保存到excel中
xls.save(file)
```

```python
import community as community_louvain
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import networkx as nx 
 
# compute the best partition
partition = community_louvain.best_partition(G)
 
# draw the graph
pos = nx.spring_layout(G)
# color the nodes according to their partition
cmap = cm.get_cmap('viridis', max(partition.values()) + 1)
nx.draw_networkx_nodes(G, pos, partition.keys(), node_size=40,
                       cmap=cmap, node_color=list(partition.values()))
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```

```python
#计算网络特征值
def graph_index(G):
    print("节点总数:", G.number_of_nodes())
    print("边总数:", G.number_of_edges())
    print("平均聚类系数(average clustering): ", nx.average_clustering(G))
    print("网络传递性(transitivity): ", nx.transitivity(G))
    print("网络直径(diameter): ", nx.diameter(G))
    print("平均最短路径(average shortest path length): ", nx.average_shortest_path_length(G))
    d = dict(nx.degree(G))
    print("平均度为：", sum(d.values())/len(G.nodes))
graph_index(sub_graph)
```
